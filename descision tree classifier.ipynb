```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(iris.data,iris.target)
#Once the classifier is created, you will feed your training data into the classifier 
#so that it can tune its internal parameters and be ready for the predictions on your future data. To tune the classifier, we run the following statement âˆ’ In : classifier.fit (X_train, Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,20))
tree.plot_tree(classifier,filled=True)
```
